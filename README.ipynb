# 1. Introduction

### 1.1. Brief Introduction of Dataset

There are several datasets provided to complete this project. I will be working with the Ford GoBike System Data. This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

This data covers the major information about the Ford GoBike System, it includes variables like; 
* duration_sec - Duration of travel in seconds
* start_station_id	- Start Station Unique Identification
* start_station_latitude - Start Station Latitude
* start_station_longitude - Start Station Longitude
* end_station_id - End Station Unique Identification
* end_station_latitude	- End Station Latitude
* end_station_longitude	- End Station Longitude
* bike_id	- Bike Unique Identification
* user_type - Bike User Types
* member_birth_year - Member Birth Year

# 2. Summary of Findings
### 2.1. Summary of findings obtained from the analysis

From my exploratory analysis, the insights obtained can be broadly classified into three;
1. Insights from Univariate Analysis
2. Insights from Bivariate Analysis
3. Insights from Multivariate Analysis

#### 1. Insights form Univariate Analysis: 
Here, there are several insights deduced and obtained. Some of which include; 
* Barplot to show the gender percentage which participated in the system. From this, Male gender recorded the highest participants, Female gender the second and the Others the third.
* Distribution Plot: This plot was used to identify the birth year range of members who participated in the system. Here, we will notice that the majority of the members are in the range of 1960 - 2000
* Pie Chart: This was used to acertain the percentage of the bike system user type. The plot shows that the 90.53% are subscribers while 9.47% are customers 
* Count plot: This was used to visualize the top twenty start station names for the trip.

#### 2. Insights form Bivariate Analysis: 

Here are the following insights were obtained;
* Scatterplot: The scatterplot was used to distance between the start lattitude and the end latitude. It shows by how far the start latitude is from the end latitude. It ascertain the claims that the start latitude is linearly correllated to the end latitude 

* Violinplot: The violinplot was used to observe the spread of the member gender with respect to the member birth year. Here, we notice that the Female gender have more aged members then the other genders.

* Hue countplot: Here, the plot shows a countplot of the member gender with respect to the user type. From this, the highest number members are male, with subscribers user type being predorminant. 

* PointPlot: The Pointplot showed the trend of members gender with respect to their birth year. There is a continuous decline in the birth year of the members.

#### 3. Insights from multivariate analysis.

* Mulivariate Scatterplot: This plot was used to estimate the duration if each bike, which indicated that most bikes completed the trip in less than 20000 secs. And younger members were predorminant in the system.

* Boxplot: Here, we realise that there were several outliers. Indicating the presence of aged members in the bike system.
